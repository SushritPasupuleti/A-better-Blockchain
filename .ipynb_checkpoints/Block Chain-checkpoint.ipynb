# A Better Blockchain

````python
import hashlib
import json
from time import time

class Blockchain(object):
    def __init__(self):
        self.chain = []
        self.current_transactions = []
        self.new_block(previous_hash=1, proof=100)

    def new_block(self, proof, previous_hash=None):
        #Create a new block, which will be added to the chain
        #proof: the proof of work algorithm

        block = {
            'index': len(self.chain) + 1,
            'timestamp': time(),
            'transactions': self.current_transactions,
            'proof': proof,
            'previous_hash': previous_hash or self.hash(self.chain[-1]),
        }
        #Empty transaction list now
        self.current_transactions = []

        self.chain.append(block)
        return block

    def new_transaction(self, sender, recipient, amount):
        #Add a transaction to the list of transactions
        #sender: Address of sender
        #recipient: Address of recipient
        #Amount: amount of currency
        self.current_transactions.append({
            'sender': sender,
            'recipient': recipient,
            'amount': amount,
        })
        return self.last_block['index'] + 1
        #Returns the index where this transaction is added.

    @staticmethod
    def hash(block):
        block_string = json.dumps(block, sort_keys=True).encode()
        return hashlib.sha256(block_string).hexdigest()

    @property
    def last_block(self):
        #returns last block of the chain
        return self.chain[-1]
````

## Proof of Work
Find a number, such that when hashed with the previous hash, it yields a hash with two zeroes.

Example Hash:
>CEE7C0E57D63484B5BD38BC5D1F1511EBE5E0906B4F7890306F5C7DA7805F382

We want it to look like:
>0ACF123F8EEEACF6190F7FC6B48FB12F9FE2F0FF57C2A4E7FC7A0D92C032E300

````python
    def proof_of_work(self, last_proof):
        proof = 0
        while self.valid_proof(last_proof, proof) is False:
            proof += 1
            
        return proof
    
    @staticmethod
    def valid_proof(last_proof, proof):
        #Validates the proof, by checking if the last 2 digits are zeroes
        trial = f'{last_proof}{proof}'.encode()
        trial_hash = hashlib.sha256(trial).hexdigest()
        
        return trial_hash[:2] == "00"
        #Returns if the proof is valid or not.
````
